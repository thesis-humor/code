# Robustness Test on Rating Inspiring - Total

```r
TedData.lm <- lm(formula = rating_inspiring_percent ~ Laughs_per_min + languages + event + Num_words_per_s 
                        + Date_2006 + Date_2007 + Date_2008 + Date_2009 + Date_2010 + Date_2011 + Date_2012 + Date_2013 + Date_2014 + Date_2015 + Date_2016,
                           data = TedTalks)
title = "Inspiring Total"

summary(TedData.lm)
```

```r
anova(TedData.lm)
```

```r
hist(x = log(TedTalks$rating_inspiring),
     main = "TedTalks",
     xlab = "Inspiring")
```

### Heteroscedasticity Test

```r
par(mfrow=c(2,2))
plot(TedData.lm, main=title)
```

```r
# Studentized Breusch-Pagan test

lmtest::bptest(TedData.lm)
```

### Autocorrelation

```r
# Breusch-Godfrey test

library(lmtest)
bgtest(TedData.lm, order = 1, order.by = NULL, type = c("Chisq", "F"),
data = list(), fill = 0)
```

```r
# Durbin-Watson test

dwtest(TedData.lm)
```

### Multicollinearity

```r
# variance inflation factor (VIF)

library(car)
car::vif(TedData.lm)
main = title
```

### Shapiro-Wilk test

```r
shapiro.test(TedTalks$Laughs_per_min)

shapiro.test(TedTalks$rating_inspiring_percent)

shapiro.test(TedTalks$rating_ingenious_percent)

shapiro.test(TedTalks$rating_unconvincing_percent)

shapiro.test(TedTalks$rating_obnoxious_percent)
```

### Heteroskedasticity Robust Standard Errors in R of Rating Inspiring

```r
library(ggplot2)
library(wooldridge)
library(lmtest)
library(dplyr)
library(lmtest)
library(sandwich)
```

```r
TedData_inspiring.lm <- lm(formula = rating_inspiring_percent ~ Laughs_per_min + languages + event + Num_words_per_s
                        + Date_2006 + Date_2007 + Date_2008 + Date_2009 + Date_2010 + Date_2011 + Date_2012 + Date_2013 + Date_2014 + Date_2015 + Date_2016,
                           data = TedTalks)

TedData_inspiring_Timing.lm <- lm(formula = rating_inspiring_percent ~ Frequency_part1 + Frequency_part2 + Frequency_part3 + languages + event + Num_words_per_s
                        + Date_2006 + Date_2007 + Date_2008 + Date_2009 + Date_2010 + Date_2011 + Date_2012 + Date_2013 + Date_2014 + Date_2015 + Date_2016,
                           data = TedTalks)
```

```r
# Rating Inspiring Total

coeftest(TedData_inspiring.lm, vcov = vcovHC(TedData_inspiring.lm, type = "HC0"))
```

```r
# Rating Inspiring Timing

coeftest(TedData_inspiring_Timing.lm, vcov = vcovHC(TedData_inspiring_Timing.lm, type = "HC0"))
```
